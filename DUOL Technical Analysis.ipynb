```python
from openbb import obb
import matplotlib.pyplot as plt
import numpy as np
```

```python
output = obb.equity.price.historical("DUOL")
```

```python
df = output.to_dataframe()
#calculate returns
df['returns'] = df['close'].pct_change()
df['log_returns'] = np.log(1+df['close'].pct_change())
df.head()
```

```python
df.plot(y = "close", figsize=(10,5), grid = True, title="DUOL Historical Stock Price")
```

```python
df.plot(y = "returns", figsize=(10,5), grid = True, title="DUOL Historical Returns")
```

```python
df.plot(y = "log_returns", figsize=(10,5), grid = True, title="DUOL Historical Returns")
```
